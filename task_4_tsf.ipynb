#Importing tha data and required libraries :

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
data=pd.read_csv('/content/drive/My Drive/datasets/Iris.csv')
data
```

```
x=data[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
y=data[["Species"]]
print(x.shape)
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=42)
```

```
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier(criterion='gini',max_depth=4,random_state=0)
tree=tree.fit(x_train, y_train)
```

```
print(tree.score(x_train, y_train))
print(tree.score(x_test, y_test))
```

#So, the accuracy of the fitted model is 99% on train dataset and 93% on test dataset

#Visualization of the decision tree model fitted :

```
from pydotplus import graph_from_dot_data
from sklearn.tree import export_graphviz
from IPython.display import Image
```

```
f_n=['SL(cm)','SW(cm)','PL(cm)','PW(cm)']
c_n=['setosa','versicolor','virginica']
dot_data=export_graphviz(tree,filled=True,rounded=True,class_names=c_n,feature_names=f_n,out_file=None)
graph=graph_from_dot_data(dot_data)
Image(graph.create_png())
```
